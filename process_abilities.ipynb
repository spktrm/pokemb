```python
import numpy as np

df = pd.read_csv("abilities.csv")
df["shortDesc embedding"] = df["shortDesc embedding"].map(
    lambda x: np.array(json.loads(x))
)
```

```python
from utils import onehot_encode

for col in {
    "effectType",
    "kind",
    "suppressWeather",
    "isBreakable",
    "isPermanent",
    "onCriticalHit",
    "condition",
}:
    df = onehot_encode(df, col)
```

```python
vectors = np.stack(df["shortDesc embedding"].values)
vectors.shape
```

```python
df = df.drop([col for col in df.columns if col.endswith(": 0")], axis=1)
```

```python
df.columns.tolist()
```

```python
etc_cols = [
    col
    for col in df.columns
    if col
    not in [
        "exists",
        "isNonstandard",
        "name",
        "num",
        "shortDesc",
        "id",
        "fullname",
        "gen",
        "desc",
        "shortDesc embedding",
    ]
]
etc_cols
```

```python
from sklearn.preprocessing import StandardScaler as Scaler
from sklearn.decomposition import KernelPCA

scaler = Scaler()
pca = KernelPCA(256, kernel="poly")

vectors = np.concatenate(
    (vectors, df[etc_cols].values.astype(float)),
    axis=-1,
)

vectors = scaler.fit_transform(pca.fit_transform(vectors)).clip(-3, 3)
vectors.shape
```

```python
vectors.shape
```

```python
ability_docs = [
    {"id": id, "name": name, "vector": vector.tolist()}
    for id, name, vector in zip(df["id"], df["name"], vectors)
]
with open("ability_documents.json", "w") as f:
    json.dump(ability_docs, f)
```

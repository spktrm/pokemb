```python
import json

with open("data.json", "r") as f:
    data = json.load(f)
```

```python
import pandas as pd
import numpy as np


from utils import (
    z_score_scale,
    onehot_encode,
    sqrt_onehot_encode,
    multihot_encode,
    json_norm_column,
    lambda_onehot_encode,
    toid,
)
```

```python
df = pd.DataFrame.from_records(data["species"])
df = df[df["isNonstandard"] != "Future"]
starting_columns = df.columns.copy()
df = df.fillna(0)
```

```python
json_norm_columns = ["baseStats", "abilities"]

for column in json_norm_columns:
    df = json_norm_column(df, column)
```

```python
onehot_encode_columns = ["gender"]
sqrt_onehot_encode_columns = []
multihot_encode_columns = ["tags", "types"]
drop_columns = [
    "exists",
    "evos",
    "genderRatio",
    "eggGroups",
    "isNonstandard",
    "tier",
    "fullname",
    "effectType",
    "kind",
    "gen",
    "shortDesc",
    "desc",
    "baseSpecies",
    "forme",
    "doublesTier",
    "canHatch",
    "unreleasedHidden",
    "maleOnlyHidden",
    "gmaxUnreleased",
    "cannotDynamax",
    "evoLevel",
    "otherFormes",
    "formeOrder",
    "canGigantamax",
    "baseForme",
    "prevo",
    "weightkg",
    "requiredItem",
    "isMega",
    "battleOnly",
    "changesFrom",
    "evoCondition",
    "evoType",
    "evoItem",
    "evoRegion",
    "evoMove",
    "cosmeticFormes",
    "requiredAbility",
    "requiredItems",
    "requiredMove",
    "isPrimal",
]

df = df.drop(drop_columns, axis=1)
```

```python
for column in onehot_encode_columns:
    df = onehot_encode(df, column)

for column in sqrt_onehot_encode_columns:
    df = sqrt_onehot_encode(df, column)

for column in multihot_encode_columns:
    df = multihot_encode(df, column)
```

```python
import math

sqrt_onehot_encode_columns = [
    "baseStats: atk",
    "baseStats: def",
    "baseStats: hp",
    "baseStats: spa",
    "baseStats: spd",
    "baseStats: spe",
    "bst",
]
z_score_columns = [
    "baseStats: atk",
    "baseStats: def",
    "baseStats: hp",
    "baseStats: spa",
    "baseStats: spd",
    "baseStats: spe",
    "bst",
    "weighthg",
]

for column in z_score_columns:
    df = z_score_scale(df, column)

for column in sqrt_onehot_encode_columns:
    df = sqrt_onehot_encode(df, column)

df = lambda_onehot_encode(df, "weighthg", lambda x: int(math.log1p(x)))
```

```python
df["abilities"] = (
    df["abilities: 0"].astype(str)
    + "\t"
    + df["abilities: 1"].astype(str)
    + "\t"
    + df["abilities: S"].astype(str)
    + "\t"
    + df["abilities: H"].astype(str)
).map(
    lambda abilities: list(
        set(
            [
                toid(ability)
                for ability in abilities.split("\t")
                if not ability.isnumeric()
            ]
        )
    )
)
```

```python
ability_df = pd.read_json("ability_documents.json")
ability_vectors = np.stack(ability_df["vector"].map(lambda x: np.array(x)).values)

from sklearn.decomposition import KernelPCA

kernelpca = KernelPCA(32)
ability_vectors = kernelpca.fit_transform(ability_vectors)
```

```python
ability_index = {
    toid(ability): index for index, ability in enumerate(ability_df["name"])
}
ability_index
```

```python
df["abilities"] = df["abilities"].map(
    lambda abilities: ability_vectors[
        [
            ability_index.get(ability)
            for ability in abilities
            if ability in ability_index
        ]
    ].mean(0)
    if [ability_index.get(ability) for ability in abilities if ability in ability_index]
    else np.zeros(ability_vectors.shape[1])
)
```

```python
df = df.drop(["abilities: 0", "abilities: 1", "abilities: S", "abilities: H"], axis=1)
```

```python
df["isNFE"] = df["nfe"]
```

```python
df = df.drop(z_score_columns, axis=1)
```

```python
vector_columns = list(sorted(set(df.columns) - set(starting_columns)))
vector_columns
```

```python
vectors = df[vector_columns].fillna(0).astype(float).values
vectors = np.concatenate((vectors, np.stack(df["abilities"].values)), axis=-1)
vectors.shape
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

pca = PCA(128)  # , kernel="rbf")
scaler = StandardScaler()

vectors = scaler.fit_transform(pca.fit_transform(vectors))
vectors = vectors.clip(-3, 3)
```

```python
from sklearn.metrics import pairwise_distances

name = "charizard"
names = df["id"].tolist()
i = names.index(name)

d = pairwise_distances(vectors, vectors)

row = d[i]
topk = 10
ind = np.argpartition(row, topk)
ind = ind[np.argsort(row[ind])]

[(names[i], row[i]) for i in ind]
```
